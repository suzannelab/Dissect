```python
## Creation d'un nouveau skelete 
## sauvegarde d'un nouveau skeleton

# liste des filaments
fil2=[]
for i,f in enumerate(skeleton.fil):
    if f.cp2.nfil >2:
        fil2.append(f)
        
# nombre de point critique au dpart, nombre de point critique ayant plus de trois filaments
#print(len(skeleton.crit), len(crit_nfil3))

#liste des points critiques
crit2=[]
for i,f in enumerate(fil2):
    crit2.append(f.cp2)
    crit2.append(f.cp1)
    
new_skeleton = Skel()
new_skeleton.crit=crit2
new_skeleton.fil=fil2

#Recupre la valeur du skelete de depart
new_skeleton.ncrit_data = skeleton.ncrit_data
new_skeleton.crit_data_name = skeleton.crit_data_name
# Faire une deep copy pour rcuprer tout les bons attributs du fichier de dpart
"""import sys
import copy
sys.setrecursionlimit(1000)
new_skeleton = copy.deepcopy(skeleton)
"""
new_skeleton.isValid()

new_skeleton.write_vtp(os.path.join(directory,"filename.vtp"))


```
